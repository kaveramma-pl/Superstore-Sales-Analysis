```python
%pip install seaborn
import seaborn as sns
```

```python
sales_['Year'] = sales_['Order Date'].dt.year  
```

```python
yearly_data = sales_.groupby(['Year', 'Category'])['Sales'].sum().round().reset_index() #yearly_data = Data Frame
```

```python
palette = {
    'Furniture': 'Black',
    'Technology': 'Yellow',
    'Office Supplies': 'orange'
}


plt.figure(figsize=(5,3))
sns.set_style('darkgrid')
sns.barplot(data=yearly_data, x='Category', y='Sales', hue='Category',  palette=palette)
plt.title ('Total Sales by Category', fontweight = 'bold')
plt.ylabel('Sales in ($)', fontsize=10, fontweight='bold')
plt.xlabel('Category', fontsize=10, fontweight='bold')


```

Technology is driving the highest sales, followed by office supllies and furniture.

```python
yearly_profit = sales_.groupby(['Year', 'Category'])['Profit'].sum().reset_index()
```

```python
palette = {
    'Furniture': 'Red',
    'Technology': 'Purple',
    'Office Supplies': 'orange'
}


plt.figure(figsize=(6,3))
sns.set_style("darkgrid")
sns.barplot(data=yearly_profit, x='Year', y='Profit', hue='Category',  palette=palette)
plt.title ('Profit by year for each category', fontweight = 'bold')
plt.legend(fontsize=8)
plt.ylabel('Profit in ($)', fontsize=10, fontweight='bold')
plt.xlabel('Year', fontsize=10, fontweight='bold')
```

Technology is giving the highest profit followed by Office supplies and furniture. Profit growth in Technology and office supplies show stable and consistent growth, whereas, furniture is underperforming and inconsistent

```python
sales_['YearMonth'] = sales_['Order Date'].dt.to_period('M').astype(str)

palette = {
    'Furniture': 'Green',
    'Technology': 'Purple',
    'Office Supplies': 'Yellow'
}

plt.figure(figsize=(12,3))

sns.lineplot(
    data=sales_,
    x='YearMonth',
    y='Sales',
    hue='Category',
    marker='o',
    estimator='sum',
    errorbar=None,
    palette=palette
)

plt.xticks(rotation=45)
plt.title('Monthly Sales Trends', fontweight='bold')
plt.legend(fontsize=8)
plt.ylabel('Sales in ($)', fontsize=10, fontweight='bold')
plt.xlabel('Year-Month', fontsize=10, fontweight='bold')
plt.tight_layout()
```

Monthly analysis reveals that Technology sales are highly volatile with sharp spikes, while Office Supplies maintains a stable and consistent growth pattern. Furniture shows moderate fluctuations. Additionally, a clear seasonality trend is observed, with higher sales toward the end of the year (Aug-Dec, many peaks as observed)

```python
yearly_data_sub = sales_.groupby(['Year','Sub-Category'])['Sales'].sum().reset_index().query(
    "`Sub-Category` in @sales_.groupby('Sub-Category')['Sales'].sum().nlargest(10).index"
)
```

```python
plt.figure(figsize=(12,4))
sns.set_style('darkgrid')

sns.barplot(
    data=yearly_data_sub,
    x='Year',
    y='Sales',
    hue='Sub-Category'
)

plt.title('Sales by Year for Top Sub-Categories', fontweight='bold')
plt.ylabel('Sales ($)', fontweight='bold')
plt.xlabel('Year', fontweight='bold')

plt.legend(fontsize=8, ncol=2)  # cleaner legend
```

Phones and Chairs bring in the most revenue and consistent growth over the years
Binders and Storage provide reliable sales, and have shown overall stability
Copiers and Machines have shown some growth, but the performance is volatile
Tables and some other sub-categorie are growing, but slower compared to the top performers

```python
top_selling_location = sales_.groupby('City')['Sales'].sum().reset_index().query(
    "City in @sales_.groupby('City')['Sales'].sum().nlargest(10).index"
)
```

```python
pivot_city = pd.pivot_table(top_selling_location, values = 'Sales', index = "City", aggfunc= sum).sort_values("Sales", ascending=False)
```

```python
pivot_city = pd.pivot_table(
    top_selling_location,
    values='Sales',
    index='City',
    aggfunc='sum'
).sort_values('Sales', ascending=False)

pivot_city = pivot_city.rename(columns={'Sales': 'Total Sales'}).reset_index()

pivot_city.style \
    .format({'Total Sales': '${:,.0f}'}) \
    .background_gradient(cmap='Spectral').set_properties(subset = "City", **{"background-color":"black"})
```

New York City is the top selling city, Los Angeles at psition 2 , Seattle at position 3, followed by San Francisco and Philadephia at 4 and 5

```python
top_selling_quantities = sales_.groupby('Sub-Category')['Quantity'].sum().nlargest(10).reset_index()
pd.pivot_table(top_selling_quantities, values = "Quantity", index = "Sub-Category", aggfunc= sum)

top_selling_quantities.style.set_properties(
    subset=['Sub-Category'],
    **{'background-color': 'black'}
).background_gradient(cmap = 'gnuplot2')
```

Binders, paper, furnishings are top 3 sub-categories sold in highest quuantities, followed by other sub-categories. 
Note - Top revenue generating sub-categories are independent of quantities
